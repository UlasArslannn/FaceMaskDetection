```python
import os 
import numpy as np 
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import cv2
from PIL import Image
from sklearn.model_selection import train_test_split
```

```python
with_mask_images = os.listdir("data/with_mask")
```

```python
without_mask_images = os.listdir("data/without_mask")
```

```python
print(with_mask_images[0:5])
```

```python
print(len(with_mask_images))
```

```python
print(len(without_mask_images))
```

```python
with_mask_images = np.array(with_mask_images)
without_mask_images = np.array(without_mask_images)
```

```python
without_mask_images.shape
```

```python
with_mask_labels  = [1] * 3720
without_mask_labels  = [0] * 3720
```

```python
with_mask_labels = np.array(with_mask_labels)
without_mask_labels = np.array(without_mask_labels)
```

```python
labels = np.concatenate((with_mask_labels, without_mask_labels))

labels.shape

print(with_mask_images.shape)

print(without_mask_images.shape)

print(with_mask_labels.shape)

print(without_mask_labels.shape)
```

```python
len(labels)

with_mask_images

with_mask_path = "data/with_mask/"
without_mask_path = "data/without_mask/"
data = []

for img_file in with_mask_images:
    image = Image.open(with_mask_path + img_file)
    image = image.resize((128,128))
    image = image.convert('RGB')
    image = np.array(image)
    data.append(image)
    
    
for img_file in without_mask_images:
    image = Image.open(without_mask_path + img_file)
    image = image.resize((128,128))
    image = image.convert('RGB')
    image = np.array(image)
    data.append(image)
```

```python

for i , image in enumerate(with_mask_images):
    ax = plt.subplot(3,4,i+1)
    img = mpimg.imread(f"data/with_mask/{with_mask_images[i]}")
    imgplot = plt.imshow(img)
    plt.axis("off")
    
    if (i == 11):
        break
plt.show()
```

```python

for i , image in enumerate(without_mask_images):
    ax = plt.subplot(3,4,i+1)
    img = mpimg.imread(f"data/without_mask/{without_mask_images[i]}")
    imgplot = plt.imshow(img)
    plt.axis("off")
    
    if (i == 11):
        break
plt.show()
```

```python
data[0].shape
```

```python
X = np.array(data)
Y = np.array(labels)
```

```python
print(X.shape)
print(Y.shape)
```

```python
X_train , X_test , Y_train , Y_test = train_test_split(X,
                                                       Y,
                                                      test_size = 0.2,
                                                      stratify=Y)
```

```python
print(X_train.shape , Y_train.shape)
```

```python
import tensorflow as tf
from tensorflow.keras.utils import normalize
```

```python
X_train = normalize(X_train, axis = -1 )
X_test = normalize(X_test , axis = -1)
```

```python
X_train
```

```python
from tensorflow.keras.models import Sequential 
from tensorflow.keras.layers import Conv2D , Dropout, Flatten
from tensorflow.keras.layers import MaxPooling2D ,Dense
```

```python
model = Sequential()
```

```python
num_of_classes = 2

model.add(Conv2D(32, kernel_size=(3,3) , activation = 'relu',
                input_shape = (128,128,3)))
model.add(MaxPooling2D(pool_size = (2,2)))

model.add(Conv2D(32, kernel_size=(3,3) , activation = 'relu'))
model.add(MaxPooling2D(pool_size = (2,2)))

model.add(Conv2D(32, kernel_size=(3,3) , activation = 'relu',))
model.add(MaxPooling2D(pool_size = (2,2)))

model.add(Flatten())
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(num_of_classes,activation = "sigmoid"))

model.compile(loss = "sparse_categorical_crossentropy", optimizer = "adam",
              metrics = ['acc'])
```

```python
history = model.fit(X_train , Y_train , validation_split=0.1, epochs = 5)
```

```python
loss , accuracy = model.evaluate(X_test , Y_test)
print('Test Accuracy =' ,accuracy)
```

```python
import cv2  # OpenCV library for resizing images

plt.figure(figsize=(15, 10))  # Adjust the figure size for a 3x5 grid

for i, image_file in enumerate(test_images_files):
    ax = plt.subplot(3, 5, i+1)
    
    # Load the image
    img = mpimg.imread(f"data/test_mask/{image_file}")
    
    # Resize the image to (128, 128)
    img_resized = cv2.resize(img, (128, 128))
    
    # Add batch dimension (1, 128, 128, 3)
    img_resized = np.expand_dims(img_resized, axis=0)
    
    # Make prediction
    pred = model.predict(img_resized)
    pred_label = np.argmax(pred)
    
    # Set the title based on the prediction
    if pred_label == 1:
        ax.set_title('Mask')
    else:
        ax.set_title('No Mask')
    
    # Display the image
    ax.imshow(img)
    ax.axis("off")  # Turn off axis

    if i == 14:  # Stop after displaying 15 images (3x5)
        break

plt.tight_layout()
plt.show()
```
